# Battleship

The code below plays a battelship game. The code flow is as follows:

    Various imports are made that will be used later.
    
    Some game constants, like the width and height of the board, the number of turns, and the length of the ship are created. They can easily be changed as the rest of the code references these constants and these constants only, so changing the value will change the whole program.
    A grid is created with the height and width previously mentioned and a nice blue background color.
    
    Some initial information is printed:
    'Let's play Battleship!'
    'You will have <num_turns> number of turns to sink a ship that is <ship_length> blocks long.'
    'Press enter to continue'
    (The press enter is done by taking in an input that will never be used, so it doesn't matter if or what the user types in there)
    
    Two functions (random_row and random_col) are defined that simply pick a row or column within the grid. These are indexed from 0 to the grid width or height respectively.
    
    Variables ship_row and ship_col are created. These form the 'head' of the ship, which is just a place for the rest of the ship to be built off of.
    
    A method (pick_ship_direction) is defined that picks a random direction (up, right, down, left) by picking an integer 0-3, each number corresponding to a direction. The ship will be built in this direction. The issue of the ship possibly going off-grid will be dealt with shortly.
    A direction is picked using the method.
    
    A method make_ship is defined which creates a list of 2 item lists. Each inner list stores the row and column of each block of the ship. This method will make sure that the ship doesn't go off the edge and that it isn't too big for the grid.
    A ship_location variable is initialized using the method.
    
    A show_ship method was created which makes all of the squares the ship holds gray (unless that square was already hit and is red). this method was mostly useful for testing purposes, but is also used if the player loses to show them where the ship was.
    
    The ship_hit and ship_sunk are defined to test if a user guess is a hit (ship_hit) and if it sinks the ship (ship_sunk). The ship is only sunk if all the squares it is on have been hit.
    
    Finally the game is actually run. A while loop goes through each turn. It has a number of parts.
        First, it prints the grid and the current turn, and then asks for the user to guess a row and column.
        Next it clears the output. That way only one grid is ever visible at a time. The program runs fast enough so that any time where nothing is being shown goes by too fast to be seen or matter.
        The game tests if the user guess was a hit (printing a message if so), and if it was it tests if the ship is sunk completely (at which point it prints a winning statement and breaks out of the while loop). It also tests if that square was already hit. To avoid issues with the ship_length being larger than the number of turns, the turn counter is NOT incremented if the ship is hit. So a user could hit the ship on their last turn and then the turn would repeat until they miss or sink the ship.
        If the user does not hit the ship, the game tests if their guess was in the ocean at all (and if it wasn't tells them to try again, NOT incrementing the turn). Then it tests if they already guessed that square (and again does NOT increment the turn if so).
        If the player simply missed the battleship, the turn IS incremented.
        If the player has used all of their turns, the ship's full location is revealed to the player and the phrase 'Game Over' is printed.
        If the game is not over, the while loop starts over.
        
The cell block directly below this is the full game including user inputs. The cell block below that uses a pre-decided location for the ship and pre-decided guesses as well as time.sleep() to show how the game runs.

```python
from random import randint
from ipythonblocks import BlockGrid
from ipythonblocks import colors
from IPython.display import clear_output

#initialize some game constants to allow for easy changing and referencing
board_width = 10
board_height = 8
num_turns = 10
#it is best if ship_length is smaller than both board_width and board_height as otherwise the ship will be cut off
#after it fills a whole column or row -- the program will not break however and the player can still win
ship_length = 5
grid = BlockGrid(board_width, board_height, fill=(123,123,234))

print 'Let\'s play Battleship!'
print 'You will have %s turns to sink a ship that is %s blocks long.\n' % (str(num_turns), str(ship_length))
pause = raw_input("Press enter to continue")

def random_row(grid):
    #Picks a random number symbolizing a row (between 0 and the height - 1)
    return randint(0, grid.height-1)

def random_col(grid):
    #Picks a random number symbolizing a row (between 0 and the width - 1)
    return randint(0, grid.width-1)

ship_row = random_row(grid)
ship_col = random_col(grid)

def pick_ship_direction():
    return randint(0,3)

ship_direction = pick_ship_direction()

#returns a list of 2 item lists that store the ship blocks in [row, column] format
def make_ship(row, col, direction):
    location = []
    for i in range(ship_length):
        if direction == 0:
            #Make sure the square actually exists on the board, otherwise move it to opposite side
            if row-i >= 0:
                location.append([row-i, col])
            else:
                #If the square does not exist, add it to the other side -- the math works out to make the other side simply i
                #First make sure the square actually exists (this will only occur if ship_size 
                #is bigger than the board_height or board_width, which means that entire row or column is filled
                #so the list location is simply returned as is, as the ship cannot grow any further)
                if i >= grid.height:
                    return location
                location.append([i, col])
        if direction == 1:
            if col+i < grid.width:
                location.append([row, col+i])
            else:
                if grid.width - 1 - i < 0:
                    return location
                location.append([row, (grid.width - 1) - i])
        if direction == 2:
            if row+i < grid.height:
                location.append([row+i, col])
            else:
                if grid.height - 1 - i < 0:
                    return location
                location.append([grid.height - 1 - i, col])
        if direction == 3:
            if col-i >= 0:
                location.append([row, col-i])
            else:
                if i >= grid.width:
                    return location
                location.append([row, i])
    return location

ship_location = make_ship(ship_row, ship_col, ship_direction)

def show_ship():
    for lst in ship_location:
        #If that spot on the ship was hit (is red) don't change the color
        if grid[lst[0], lst[1]].red != 255:
            grid[lst[0], lst[1]] = colors['Gray']
            
def ship_hit(row, col):
    for lst in ship_location:
        if [row, col] == lst:
            return True
    return False

def ship_sunk():
    for lst in ship_location:
        #128 is the rgb value for 'Gray', 123 is the rgb value for the light blue being used
        #if any block is one of those two colors, the ship is not sunk
        if (grid[lst[0], lst[1]].red == 128) or (grid[lst[0], lst[1]].red == 123):
            return False
    return True

#for testing purposes to check when battleship is hit, comment out in final product
#print ship_row + 1
#print ship_col + 1
#show_ship()
#grid.show()

turn = 1
while turn <= num_turns:
    grid.show()
    print 'Turn', turn
    
    #Initializes guess_row to a valid number by taking input from user
    while True:
        try:
            #Take input from user (in range 1 to board_height) and subtract one
            #in order to set the guesses starting at zero to match index
            guess_row = int(raw_input("Guess Row (1-%s): " % str(grid.height))) - 1
            #If the entered row number is not a valid row, raise ValueError so it will ask again
            if guess_row < 0 or guess_row >= grid.height: raise ValueError
            break
        except ValueError:
            print "That was not a valid number.  Please try again."
    
    #Initializes guess_col to a valid number by taking input from user
    while True:
        try:
            #Take input from user (in range 1 to board_width) and subtract one
            #in order to set the guesses starting at zero to match index
            guess_col = int(raw_input("Guess Row (1-%s): " % str(grid.width))) - 1
            #If the entered column number is not a valid column, raises ValueError so it will ask again
            if guess_col < 0 or guess_col >= grid.width: raise ValueError
            break
        except ValueError:
            print "That was not a valid number.  Please try again."
    
    clear_output()
    
    if ship_hit(guess_row, guess_col):
        #Test if the square is already red (already hit)
        #if not, they got a new hit
        if grid[guess_row, guess_col].red == 255:
            print 'You already hit me there. No repeats! I\'ll give you that turn back, however.'
        else:
            grid[guess_row, guess_col] = colors['Red']
            if ship_sunk():
                grid.show()
                print "Arg, you got me! My battleship is sunk.\nCongratulations, you win!"
                break
            print "Hit! You got a hit! Have an extra turn..."
    else:
        #a grid square will only have the red of it's rgb value set to zero if it is black
        #tests if the square is black to see if they already guessed that location
        #gives them their turn back if they have
        elif(grid[guess_row, guess_col].red == 0):
            print "You guessed that one already. Let's retry that."
        else:
            print "You missed my battleship!"
            turn += 1
            grid[guess_row, guess_col] = colors['Black']
        if turn > num_turns:
            show_ship()
            grid.show()
            print 'Game Over'
            break
```
