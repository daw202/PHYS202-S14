For this question refer to the solutions for the simple pendulum we got using scipy's odeint from the beginning of the ODE Tour.

**(a)** Use the numerical solution to see what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs. t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 or π ?

```python
from scipy.integrate import odeint
```

```python
g = 9.8 #m/s^2
ell = 0.5 #m

def deriv(theta_current, t):  # return derivatives of the array theta
    g = 9.8 #m/s^2
    ell = 0.5 #m
    theta = theta_current[0]
    omega = theta_current[1]
    dtheta = omega
    domega = -(g / ell) * np.sin(theta)
    return np.array([ dtheta, domega ])

t = np.arange(0., 10.+.01, 0.01)

theta_inits = []
for i in range(5):
    theta_inits.append([i * pi / 4, 0])
theta_inits = np.array(theta_inits)

#pass the function deriv to the solver with initial values and integration variable
thetas = []
for i in theta_inits:
    thetas.append(odeint(deriv, i, t))
thetas = np.array(thetas)
```

```python
plt.plot(t, thetas[0][:,0], label = 'theta=0')
plt.plot(t, thetas[1][:,0], label = 'theta=pi/4')
plt.plot(t, thetas[2][:,0], label = 'theta=pi/2')
plt.plot(t, thetas[3][:,0], label = 'theta=3pi/4')
plt.plot(t, thetas[4][:,0], label = 'theta=pi')

plt.xlabel(r'$t (seconds)$', fontsize = 20)
plt.ylabel(r'$\theta (rad)$', fontsize = 20)
plt.ylim(-3, 10)
plt.grid()
plt.legend()
plt.show()
```

### Question 2

**(a)** Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.  
dy/dx=x+2y  
Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

```python
def Euler(step):
    xmax = 1.+step
    dx = step
    nsteps = int(xmax / dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    #y[0] = 0. #commented because it's already a zero - reduntant
    
    for i in range(len(x) - 1):
        x[i+1] = x[i] + dx
        y[i+1] = y[i] + dx * (x[i] + 2*y[i])
    
    return x, y

x1, y1 = Euler(.25)
x2, y2 = Euler(.02)
```

**(c)** Now solve the same problem using Euler-Cromer and the Midpoint method for the same pair of step sizes and plot the results along with the exact solution. Then also plot the difference between the approximate numerical solutions and the exact solution. Which combination of method and step-size gives the best result?

```python
def EulerCromer(step):
    xmax = 1.+step
    dx = step
    nsteps = int(xmax / dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    #y[0] = 0. #commented because it's already a zero - reduntant
    
    for i in range(len(x) - 1):
        x[i+1] = x[i] + dx
        y[i+1] = y[i] + dx * (x[i+1] + 2*y[i])
    
    return x, y

xc1, yc1 = EulerCromer(.25)
xc2, yc2 = EulerCromer(.02)
print yc1[1/.25]
print yc2[1/.02]
```

```python
plt.plot(xc1, yc1)
plt.plot(xc2, yc2)
x = np.arange(0, 1.01, .01)
y = .25 * exp(2 * x) - 0.5 * x - 0.25
plt.plot(x, y)
plt.xlim(0, 1)
plt.show()
```

```python
def EulerMidpoint(step):
    xmax = 1.+step
    dx = step
    nsteps = int(xmax / dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    
    #y[0] = 0. #commented because it's already a zero - reduntant
    
    for i in range(len(x) - 1):
        x[i+1] = x[i] + dx
        y[i+1] = y[i] + dx * ((x[i+1] + x[i])/2. + 2*y[i])
    
    return x, y

xm1, ym1 = EulerMidpoint(.25)
xm2, ym2 = EulerMidpoint(.02)
print ym1[1/.25]
print ym2[1/.02]
```

```python
plt.plot(xm1, ym1)
plt.plot(xm2, ym2)
x = np.arange(0, 1.01, .01)
y = .25 * exp(2 * x) - 0.5 * x - 0.25
plt.plot(x, y)
plt.xlim(0, 1)
plt.show()
```

### Question 3

We can use numerical solvers and widgets to explore the dynamics of a projectile that is subject to a drag force proportional to the square of its velocity - "quadratic drag".

**(a)** Create a derivative function to solve this system of equations and use odeint to solve it for a projectile fired at an angle of 50 degrees from the horizontal at an initial speed of 30 m/s. Track the projectile's motion over the time interval of 0 to 200 seconds and let (x0,y0) = (0,0). You'll have to calculate (vx0,vy0) from the information provided. Let vter=44 m/s. Plot the trajectory of the projectile (y vs. x) to see how far it can travel in the x direction before landing.

```python
theta0 = 50. * (2 * pi) / 360
v0 = 30.
tf = 200.
t = np.arange(0., tf + .01, .01)
x0 = 0.
y0 = 0.
vter = 44.
vx0 = v0 * np.cos(theta0)
vy0 = v0 * np.sin(theta0)

def derivfunc(values_current, t):
    '''values_current takes form [x, y, vx, vy, vter]'''
    x = values_current[0]
    y = values_current[1]
    vx = values_current[2]
    vy = values_current[3]
    vter = values_current[4]
    
    dx = vx
    dy = vy
    dvx = - g * vx/vter**2 * np.sqrt(vx**2 + vy**2)
    dvy = - g * (1 + vy/vter**2 * np.sqrt(vx**2 + vy**2))
    return np.array([dx, dy, dvx, dvy, vter])

init_values = np.array([x0, y0, vx0, vy0, vter])
projectile = odeint(derivfunc, init_values, t)

below_zero = 0
for i in range(len(projectile[:,1])):
    if projectile[:,1][i] < 0:
        below_zero = i
        break

plt.plot(projectile[:,0], projectile[:,1])
plt.ylim(0, numpy.amax(projectile[:,1] * 1.4))
plt.xlim(0, projectile[:,0][below_zero])
plt.show()
```

(b) Now create an interact object that uses sliders for the initial speed, terminal velocity, launch angle, and time. Have it plot two curves - one representing the solution with quadratic drag and the other in the absence of drag:

x(t)=x0+vx0t

y(t)=y0+vy0t−1/2gt2

```python
from IPython.html.widgets import interact, interactive
```

```python
def projectile_motion(v0, vter, theta0, tf):
    theta0 *= (2 * pi) / 360
    t = np.arange(0, tf + .01, .01)
    x0 = 0.
    y0 = 0.
    vx0 = v0 * np.cos(theta0)
    vy0 = v0 * np.sin(theta0)
    
    xnodrag = vx0 * t + x0
    ynodrag = y0 + vy0 * t - 0.5 * g * t * t
    plt.plot(xnodrag, ynodrag, label = 'Without drag')

    init_values = np.array([x0, y0, vx0, vy0, vter])
    projectile = odeint(derivfunc, init_values, t)

    below_zero = 0
    for i in range(len(projectile[:,1])):
        if projectile[:,1][i] < 0:
            below_zero = i
            break
    
    maxy = numpy.amax(numpy.array([numpy.amax(projectile[:,1]), numpy.amax(ynodrag)]))
    plt.plot(projectile[:,0], projectile[:,1], label = 'With quadratic drag')
    plt.ylim(numpy.amin(projectile[:,1]), maxy * 1.4)
    plt.xlim(numpy.amin(projectile[:,0]), numpy.amax(projectile[:,0]) * 1.1)
    plt.legend()
    plt.show()

v = interact(projectile_motion, v0 = (0., 100.), vter = (10., 107.), theta0 = (0., 90.), tf =(0., 20.))
```
