**(b)** Next write out a new file named "clean_data.dat". The new file should contain the same data as the old file, but with uniform formatting and aligned columns.

```python
f = open('clean_data.dat', 'w')
for i in range(len(extract_rows)):
    for j in range(len(extract_rows[i])):
        f.write(str(extract_rows[i,j]))
        if len(str(extract_rows[i,j])) <= 7:
            f.write('\t')
        f.write('\t')
    f.write('\n')
f.close()

f = open('clean_data.dat', 'r')
print f.read()
f.close()
```
