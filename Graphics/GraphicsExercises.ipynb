```python
"""
Demo of the `streamplot` function.

A streamplot, or streamline plot, is used to display 2D vector fields. This
example shows a few features of the stream plot function:

    * Varying the color along a streamline.
    * Varying the density of streamlines.
    * Varying the line width along a stream line.
"""
import numpy as np
import matplotlib.pyplot as plt

Y, X = np.mgrid[-10:10:100j, -10:10:100j]
U = -1 - X + Y
V = 1 + X**2 - Y**2
speed = np.sqrt(U*U + V*V)

plt.streamplot(X, Y, U, V, color=U, linewidth=2, cmap=plt.cm.RdPu)
plt.colorbar()

f, (ax1, ax2) = plt.subplots(ncols=2)
ax1.streamplot(X, Y, U, V, density=[3, 3])

lw = 5*speed/speed.max()
ax2.streamplot(X, Y, U, V, density=0.9, color='m', linewidth=lw)

plt.show()
```

**5.** Determine the electric potential for a quadrupole consisting of

q1 = -1.0 nC at (-2 m, 0 m)  
q2 = -1.0 nC at (+2 m, 0 m)  
q3 = +1.0 nC at (0 m, -2 m)  
q4 = +1.0 nC at (0 m, +2 m)  


**(a)** Make a surface plot on the range -4 <(x,y)< 4 using 101 bins in x and 101 bins in y. In this case, do not let k = q = d = 1. Use dimensionful quantities and compute the value of the potential in volts.
Recall k = 8.9875517873681764 × 109 N⋅ m2/C2.

```python
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1, projection = '3d')
X = np.arange(-4, 4, 8/101.0)
Y = np.arange(-4, 4, 8/101.0)
X, Y = np.meshgrid(X, Y)

#information on charges stored in [x, y, charge in nC] format
charges = {'q1': [-2, 0, -1.0 * 10**-9], 'q2': [2, 0, -1.0 * 10**-9], 'q3': [0, -2, 1.0 * 10**-9], 'q4': [0, 2, 1.0 * 10**-9]}

K = 8.9875517873681764 * 10**9

Vone = K*charges['q1'][2]/(np.sqrt((X-charges['q1'][0])**2 + (Y-charges['q1'][1])**2))
Vtwo = K*charges['q2'][2]/(np.sqrt((X-charges['q2'][0])**2 + (Y-charges['q2'][1])**2))
Vthree = K*charges['q3'][2]/(np.sqrt((X-charges['q3'][0])**2 + (Y-charges['q3'][1])**2))
Vfour = K*charges['q4'][2]/(np.sqrt((X-charges['q4'][0])**2 + (Y-charges['q4'][1])**2))

Z = Vone + Vtwo + Vthree + Vfour

surf = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.seismic,
        linewidth=0, antialiased=True)
```

**(b)** Now make a contour plot of the same data, with appropriate contour levels to see the shape of the potential.

```python
levels = np.array([-150, -100, -50, -10, -1, 0, 1, 10, 50, 100, 150])
CS = plt.contour(X, Y, Z,levels,linewidths=2,cmap=cm.coolwarm)
plt.clabel(CS, inline=1, fontsize=10)
```
